```python
from bs4 import BeautifulSoup
import requests, sys, os
import json
from collections import defaultdict
```

```python
base_url = "http://lotr.wikia.com"
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36"

char_dict = {}

def getCharacterInfo(name, char_url):
    char_info = {}
    char_info["Name"] = name 
    response = requests.get(char_url)
    html = response.text.encode("utf-8")
    soup = BeautifulSoup(html, "html5lib")
    char_info_section = soup.findAll('div', {"class":"pi-item pi-data pi-item-spacing pi-border-color"})
    for i in char_info_section:
        h3_val = i.find('h3').text
        if h3_val == "Birth" or h3_val == "Death":
            try:
                div_val = i.find('a').text
            except:
                div_val = i.find('div').text
        elif h3_val == "Other Names":
            continue

# The commented out else statement works for other names but not culture. Since other names isn't important I decided to forgo it
#         else:
#             #Handle line breaks (<br>) in character's information turning it into a comma separated string
#             if i.find('div').find_all('br') and h3_val != "Culture":
#                 div_val = []
#                 for br in i.find('div').find_all('br'):
#                     if div_val == []:
#                         div_val += [br.previousSibling]
#                     div_val += [br.nextSibling]
#                 print(div_val)
#                 div_val = ",".join(div_val)

        else:
            div_val = i.find('div').text

        char_info[h3_val] = div_val
    return(char_info)
    

def saveCharacters():
    outfile = ('Characters.csv', 'w')
    next_url = base_url + "/wiki/Category:Characters?pageuntil=Borlach"

    while(next_url):
        response = requests.get(next_url)
        #If the next_url isn'tset in the for loop, there are no more pages and the while loop exits
        next_url = ""
        html = response.text.encode('utf-8')
        soup = BeautifulSoup(html, "html5lib")
        #Section with char list
        char_section = soup.findAll('div', {"id": "mw-pages"})
        #Find all the links in the character table section. This includes the next & previous page links
        char_lst = char_section[0].find_all('a')
        for char_link in char_lst:
            if char_link.text == "next 200":
                next_url = base_url + char_link['href']
            elif char_link.text == "previous 200":
                continue
            else:
                char_name = char_link.text
                char_url = base_url + char_link['href']
                info = getCharacterInfo(char_name, char_url)
                char_dict[char_url] = info
    return(char_dict)

#Gets the number of characters in the wiki overall
def numberOfCharacters(url):    
    response = requests.get(url)
    html = response.text.encode('utf-8')

    soup = BeautifulSoup(html, "html5lib")

    #Section with char list
    char_section = soup.findAll('div', {"id": "mw-pages"})

    # char_section is result set with one element. The first p element contains a sentence about the number of characters as such
    # "The following 200 pages are in this category, out of 939 total."
    num_chars = [i for i in char_section[0].find('p').text.split() if i.isdigit()][-1]
    return(num_chars)

print(saveCharacters())
```
